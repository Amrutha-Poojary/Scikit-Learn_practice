```python
# Value imputation
from sklearn.impute import SimpleImputer
```

```python
print(f"number of missing values in age column is {df['age'].isnull().sum()}")
```

```python
imp = SimpleImputer(strategy='mean')
df['age'] = imp.fit_transform(df[['age']])
print(f"number of null values in age column after inputation:  {df['age'].isnull().sum()}")
```

```python
def get_parameters(df):
  Parameters = {}
  for col in df.columns[df.isnull().any()]:
    if df[col].dtype =='float64' or df[col].dtype =='int64':
      strategy ='mean'
    else:
      strategy ='most_frequent'

    missing_Values = df[col].isnull().sum()
    Parameters[col] = {'missing_values': missing_Values, 'strategy': strategy}
  return Parameters

get_parameters(df)
```

```python
parameters = get_parameters(df)
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
import pandas as pd

for col, parm in parameters.items():
  strategy = parm['strategy']
  imp = SimpleImputer(strategy=strategy)
  # Fit and transform the column.
  imputed_data = imp.fit_transform(df[[col]])

  # For non-numeric columns with 'most_frequent', convert back to original dtype.
  if (df[col].dtype == 'object' or df[col].dtype.name == 'category') and strategy == 'most_frequent':
      # Convert the imputed data back to a pandas Series with the original dtype
      original_dtype = df[col].dtype
      df[col] = pd.Series(imputed_data.flatten(), index=df.index).astype(original_dtype)
  else:
      df[col] = imputed_data
```
